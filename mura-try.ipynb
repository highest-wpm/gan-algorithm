```python
import os
import cv2
import matplotlib.pyplot as plt
img = cv2.imread(r'C:\img\mura\Pin\20180213_165758_ILDM04_F882C039CA_15203_T_1_#15_3.jpg')
img_inhance=img*3+20
fig = plt.figure()
plt.imshow(img_inhance)
```

```python
bluravg= cv2.blur(img_inhance, (9, 9))
fig = plt.figure()
plt.imshow(bluravg)
```

```python
blurred = cv2.GaussianBlur(img_inhance, (3, 3), 0)
canny = cv2.Canny(blurred, 30, 150)
fig = plt.figure()
plt.imshow(cv2.cvtColor(canny, cv2.COLOR_GRAY2RGB))
```
