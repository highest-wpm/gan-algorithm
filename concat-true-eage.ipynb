```python
import os
import cv2
import numpy as np
imgnamelist=os.listdir(r'C:\img\musk\musk-true')
for i in imgnamelist:
    img = cv2.imread(r'C:\img\musk\musk-true\{}'.format(i))
    gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    canny = cv2.Canny(blurred, 30, 150)
    canny_bgr=cv2.cvtColor(canny, cv2.COLOR_GRAY2BGR)
    concat=np.hstack([img,canny_bgr])
    cv2.imwrite(r'C:\img\musk\musk-concat\concat-{}'.format(i), concat)
```
