```python
import os
import cv2
imgnamelist=os.listdir(r'C:\img\musk')
for i in imgnamelist:
    gray = cv2.imread(r'C:\img\musk\{}'.format(i), cv2.IMREAD_GRAYSCALE)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    canny = cv2.Canny(blurred, 30, 150)
    cv2.imwrite(r'C:\img\musk-Eage\Eage-{}'.format(i), canny)
```
